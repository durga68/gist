### Filtering Arrays

```python
import numpy as np
ary=np.array([11,12,13,14])
x=[True,False,True,False]
nary=ary[x]
print(nary)
```

```python
import numpy as np
ary=np.array([11,12,13,14])
print(ary)
x=[True,False,True,False]
nary=ary[x]
print(nary)
```

```python
ar=np.array([1,2,3,4,5])
filter_ar=ar%2==0
new_ar=ar[filter_ar]
print(filter_ar)
print(new_ar)
```

```python
ar=np.array([1,2,3,4,5])
filter_ar=ar%2!=0
new_ar=ar[filter_ar]
print(filter_ar)
print(new_ar)
```

```python
ar=np.array([1,2,3,4,5])
filter_ar=ar%2==1
new_ar=ar[filter_ar]
print(filter_ar)
print(new_ar)
```
